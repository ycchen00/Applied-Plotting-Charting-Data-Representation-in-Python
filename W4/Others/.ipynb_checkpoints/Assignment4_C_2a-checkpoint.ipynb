```python
# Becoming Independent Data Scientist
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook

# Removes default white gridlines from the plot:
plt.rcParams["axes.grid"] = False
```

```python
# Read and clean data for manufacturing employees:
manuf = pd.read_csv('final1.csv')
manuf.head()
```

```python
from collections import ChainMap
#manuf=manuf.rename(columns={'Employment (1)':'Employment'}, {'Annual mean wage (2)': 'Annual mean wage'})
#manuf = manuf.rename(columns={'old_name':new})
manuf=manuf.rename(ChainMap({'Employment (1)':'Employment'}, {'Annual mean wage (2)': 'Annual mean wage'}), axis='columns')
manuf
```

```python
# Read and clean data for employees in professional and business services:
prof_business = pd.read_csv('final22.csv')
#prof_business['DATE'] = pd.to_datetime(manuf['DATE'])
prof_business
```

```python
from collections import ChainMap
prof_business=prof_business.rename(ChainMap({'Annual mean wage (2)': 'Annual mean wage'}), axis='columns')
prof_business
```

```python
# Read and clean data for average weekly wages for employees in private organisations:
avg_weekly = pd.read_csv('final33.csv')
#avg_weekly['DATE'] = pd.to_datetime(avg_weekly['DATE'])
avg_weekly
```

```python
from collections import ChainMap
#manuf=manuf.rename(columns={'Employment (1)':'Employment'}, {'Annual mean wage (2)': 'Annual mean wage'})
#manuf = manuf.rename(columns={'old_name':new})
avg_weekly =avg_weekly .rename(ChainMap({'Employment (1)':'Employment'}, {'Annual mean wage (2)': 'Annual mean wage'}), axis='columns')
avg_weekly 
```

```python
# Monthly date range:
#dates = pd.date_range(start='1/1/1990', end='1/2/2019', freq='MS')

fig, ax1 = plt.subplots()
ax1.plot(manuf['Hourly mean wage'],manuf['Percent of industry employment'], label=' levels of employment')
ax1.plot(prof_business['Hourly mean wage'], prof_business['Percent of industry employment'], label='concentration of employment')
ax1.set_xlabel('Hourly mean wage')
ax1.set_ylabel('Industry employment rate in %')
ax1.tick_params('y')

# frameon turns the frame around the legend on. framealpha adjusts the transparency of the box.


ax2 = ax1.twinx()
ax2.plot(avg_weekly['Hourly mean wage'],avg_weekly['Percent of industry employment'], 'r', label='Top paying industries')
#ax2.set_ylabel('Hourly mean wage')
ax2.tick_params('y')
ax2.legend(loc='upper left', fontsize=9, frameon=True, framealpha=1)
ax1.legend(loc='upper right', fontsize=9, frameon=True, framealpha=1)
plt.title('Occupational Employment and Wages in Ann Arbor, Michigan')
fig.tight_layout()
plt.show()
```
